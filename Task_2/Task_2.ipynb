Unemployment is measured by the unemployment rate which is the number of people
who are unemployed as a percentage of the total labour force. We have seen a sharp
increase in the unemployment rate during Covid-19, so analyzing the unemployment rate
can be a good data science project.

```python
colors = sns.color_palette("husl", len(df['Region'].unique()))
np.random.shuffle(colors)

plt.figure(figsize=(15, 10))
sns.countplot(y="Region", data=df, palette=colors)
plt.show()
```

```python


data = {
    'Estimated Unemployment Rate (%)': [5.2, 4.8, 6.1, 5.5],
    'Estimated Employed': [2000, 2500, 1800, 2200],
    'Estimated Labour Participation Rate (%)': [60, 62, 58, 61]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
corr = df[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(corr, annot=True, linewidths=0.5, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
